# Pendulum

```python
A = np.array([[0.,1.],[0.,0.]])

def b(t,x):
    return np.array([0,-9.81/1*np.sin(x[0])])

x0 = np.array([np.pi-1,0])
t0 = 0
h = 0.01
N_steps = 1000

Pendulum = ODE_Solver(A,b)
x,t = Pendulum.run_RK4(x0,t0,h,N_steps)
x2,t2 = Pendulum.run_Explicit_mid(x0,t0,h,N_steps)

plt.plot(t,x[:,0])
plt.plot(t2,x2[:,0])
plt.show()
```

# Dissipative Driven Harmonic Oscillator

```python
w0 = 1
w = 0
eta = 0.002

x0 = np.array([np.pi-0.01,0])
t0 = 0
h = 0.01
N_steps = 4000

A = np.array([[0.,1.],[-w0**(2),-2*eta*w0]])

def b(t,x):
    return np.array([0., 5/3*np.sin(w*t)])

Harmonic_Driven = ODE_Solver(A,b)
x,t = Harmonic_Driven.run_RK4(x0,t0,h,N_steps)

plt.plot(t,x[:,0])
```

# Lotka-Volterra

```python
alpha = 0.1
beta = 0.01
delta = 0.002
gamma = 0.2

A = np.array([[alpha,0],[0,-gamma]])

def b(t,x):
    return np.array([-beta*x[0]*x[1],delta*x[0]*x[1]])


x0 = np.array([10,1])
t0 = 0
h = 0.01
N_steps = 20000

LV = ODE_Solver(A,b)

x,t = LV.run_RK4(x0,t0,h,N_steps)

plt.plot(t,x[:,0],label="prey")
plt.plot(t,x[:,1],label="predator")
plt.legend()
plt.show()
```
